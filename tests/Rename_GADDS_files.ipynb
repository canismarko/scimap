```python
import os
import re
import shutil

DIR = 'test-data-xrd/xrd-map-gadds/'
```

```python
map_re = re.compile("map-([a-z0-9]+)()\.(plt)")
# map_re = re.compile("map-([a-z0-9]+)(_01_000)\.(gfrm)")
# map_re = re.compile('map-([a-z0-9]+)(_01)\.(jpg)')
# map_re = re.compile('map-([a-z0-9]+)(01)\.(p4p)')
for file in os.listdir(DIR):
    match = map_re.match(file)
    if match:
        pos, frame, ext = match.groups()
        pos = int(pos, 16)
        new_file = "map-{pos}{frame}.{ext}".format(pos=pos, frame=frame, ext=ext)
        print(file + " -> " + new_file)
        shutil.move(os.path.join(DIR, file),
                    os.path.join('test-data-xrd/xrd-map-gadds-new', new_file))
```
